```python
import json
import csv
import pandas as pd, numpy as np
```

```python
# Open the JSON file
with open('PRD-DEPLOY.json') as f:
    data = json.load(f)

# Print the JSON data
print(data) 

# Print JSON with indentation
print(json.dumps(data, indent=4))
```

```python
# Normalize the JSON data into a dataframe
df = pd.json_normalize(data)

# Print the dataframe
df.head()
```

```python
# Extract column names
column_names = df.columns.tolist()

# Create a CSV file with column names
with open('column_names.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(column_names)

# Load up the CSV file
df = pd.read_csv('column_names.csv')
df.count()
```
